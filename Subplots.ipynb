```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline
```

```python
X = np.linspace(0,10, 11)
y = X**4
```

```python
X
```

```python
y
```

```python
a = np.arange(0,10)
b = a*2
```

```python
a
```

```python
b
```

```python
fig, axes = plt.subplots()

axes.plot(a, b, 'b')
axes.set_xlabel('X')
axes.set_ylabel('y')
axes.set_title('Blue Line')

plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
axes.shape
```

```python
fig
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for axe in axes:
    axe.plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(a,b)
axes[1].plot(X,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(a,b)
axes[1][1].plot(X,y)

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8))


axes[0][0].plot(a,b)
axes[0][0].set_title('0 0 Title')


axes[1][1].plot(X,y)
axes[1][1].set_title('1 1 Title')
axes[1][1].set_xlabel('1 1 X Label')

axes[0][1].plot(y,X)
axes[1][0].plot(b,a)


fig.suptitle("Figure Level",fontsize=16)


plt.show()
```
